### Importar librerías

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Creamos datos aleatorios

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
y
```

### Gráfica en función de Y

```python
plt.plot(x,y)
plt.show()
```

#### Cambiando color a la gráfica

```python
plt.plot(x,y,'m')
plt.show()
```

#### Colocando marcadores en forma de círculos

```python
plt.plot(x,y,'mo')
plt.show()
```

#### Colocando marcadores en forma de X

```python
plt.plot(x,y,'rx')
plt.show()
```

#### Colocando marcadores en forma de cuadrados con línea contínua

```python
plt.plot(x,y,'bs-')
plt.show()
```

#### Colocando marcadores en forma de diamantes con línea punteada

```python
plt.plot(x,y,'yD:')
plt.show()
```

### Gráfica Histograma

```python
plt.hist(y)
plt.show()
```

### Gráfica de Pie

```python
plt.pie(x)
plt.show()
```

### Correlación entre variables

```python
plt.scatter(x,y)  #correlacion entre la variable x e y
plt.show()
```

### Distribución de Datos

```python
plt.boxplot(x)
plt.show()
```
