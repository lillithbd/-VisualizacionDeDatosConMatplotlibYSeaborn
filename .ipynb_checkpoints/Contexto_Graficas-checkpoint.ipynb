### Importar librerías

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Crear Set de Datos

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

### Gráfico

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

### Contexto a Gráficos con Título

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title('Relación X-Y')

axes[1].plot(y,x)
axes[1].set_title('Relación Y-X')
```

### Contexto agregando etiquetas

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title('Relación X-Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')


axes[1].plot(y,x)
axes[1].set_title('Relación Y-X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

### Cambiar tamaño a figuras

```python
fig, axes = plt.subplots(1,2, figsize=(5,5))
axes[0].plot(x,y)
axes[0].set_title('Relación X-Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')


axes[1].plot(y,x)
axes[1].set_title('Relación Y-X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

### Agregando etiquetas

```python
fig, axes = plt.subplots(1,2, figsize=(5,5))
axes[0].plot(x,y, label="$sin(x)$")
axes[0].set_title('Relación X-Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()


axes[1].plot(y,x)
axes[1].set_title('Relación Y-X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

### Aplicar etiquetas sin usar orientado a objetos

```python
plt.plot(x,y,label='sin(x)')
plt.title('Este es un título')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='lower left', bbox_to_anchor=(1,0.9))
```
