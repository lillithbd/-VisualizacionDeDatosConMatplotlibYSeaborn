### Importar librerías

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Crear Set de Datos

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

 ### Crear primero una figura que será el lienzo de nuestras gráficas

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

### Creando varias gráficas

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

### Acceder a los gráficos mediante el nombre almacenado en una tupla.

```python
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```

### Generar una matriz tipo fig, con 2 filas y 4 columnas

```python
fig, (ax1,ax2) = plt.subplots(2,4)
#ax1.plot(x,y,'b')
#ax2.plot(y,x,'r')
```

### Como acceder a estas filas y estas columnas

```python
fig, axes = plt.subplots(2,4)
axes[0,0].plot(x,np.cos(x))
axes[0,1].plot(x,np.sin(x),'y')
axes[0,2].plot(x,np.tan(x),'r' )
axes[0,3].plot(x,np.cos(x)**2)
```

### Alternativa para generar una matriz de tipo fig, accediendo a cada gráfica mediante los nombres dados en una tupla

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2,4)
ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x),'y')
ax3.plot(x,np.tan(x),'r' )
ax4.plot(x,np.cos(x)**2)
ax5.plot(x,np.cos(x)/np.sin(x)) #cotangente
ax6.plot(x,1/np.cos(x),'m') #secante
ax7.plot(x,1/np.sin(x),'b' ) #cosecante
ax8.plot(x,np.tan(x)**2,'g')
fig.tight_layout()
```
