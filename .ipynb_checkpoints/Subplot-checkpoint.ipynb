### Importar librer√≠as

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Crear Set de Datos

```python
x = np.linspace(0,5,11)
y = x ** 2
```

### Subplot

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r--')
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

### Agregar diferentes Plots al gr√°fico

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r--') # Plot 1
plt.plot(y,x, 'b:') #Plot 2

plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

### Invertir la matriz de los gr√°ficos

```python
plt.subplot(2,1,1) #Grafico 1
plt.plot(x,y,'r--') #Plot 1
plt.plot(y,x,'b:')  #Plot 2

plt.subplot(2,1,2) #Grafico 2
plt.pie(y)         #Plot 1  
plt.show()
```
