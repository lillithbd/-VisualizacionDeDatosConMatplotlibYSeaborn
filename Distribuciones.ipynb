### Importando librer√≠as

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Cargando datos

```python
tips = sns.load_dataset('tips')
tips.head()
```

### Creaundo histograma 

```python
sns.histplot(data=tips, x='tip')
plt.show()
```

### Especificando cuantas barras quiero para esta frecuencia

```python
sns.histplot(data=tips, x='tip', bins=15)
plt.show()
```

### Acumulando valores

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=True)
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex')
plt.show()
```

### Stat 

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='frequency')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='probability')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='percent')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='density')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count')
plt.show()
```

### Cambiar c√≥mo quiero presentar mis datos con multple

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='stack')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='layer')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='dodge')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='fill')
plt.show()
```

### Gr√°fico kde (gr√°fico de densidad)

```python
sns.kdeplot(data=tips, x='tip',hue='sex')
plt.show()
```

### Cumulative

```python
sns.kdeplot(data=tips, x='tip',hue='sex', cumulative=False)
plt.show()
```

### Dibuja √°rea bajo la curva

```python
sns.kdeplot(data=tips, x='tip',hue='sex', cumulative=False, shade=True)
plt.show()
```

### Ajustar la gr√°fica y los par√°metros

```python
sns.kdeplot(data=tips, x='tip',hue='sex', cumulative=False, shade=True,bw_adjust=1)
plt.show()
```

### Gr√°fico ecdfplot

```python
sns.ecdfplot(data=tips,x='tip',hue='sex') #Gr√°fico escalonado de las proporcion de las propinas a lo largo de su frecuencia
plt.show()
```

```python
sns.ecdfplot(data=tips,x='tip',hue='sex',stat='count')
plt.show()
```

### Diagrama de distribucion displot

```python
sns.displot(data=tips,x='tip',hue='sex')
plt.show()
```

```python
sns.displot(data=tips,x='tip',hue='sex', kind='kde')#cambiando tipo de grafico
plt.show()
```

```python
sns.displot(data=tips,x='tip',hue='sex', kind='hist')#cambiando tipo de grafico
plt.show()
```

```python
sns.displot(data=tips,x='tip',hue='sex', kind='hist', multiple='stack')#cambiando tipo de grafico
plt.show()
```
