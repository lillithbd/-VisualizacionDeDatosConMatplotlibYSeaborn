### Importando librer√≠as

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Cargando datos

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
#Llevar los datos a una estructura matricial especial
tips.corr(numeric_only=True)
```

### Heatmap

```python
sns.heatmap(tips.corr(numeric_only=True))
plt.show()
```

```python
#adicionando valores de la correlacion
sns.heatmap(tips.corr(numeric_only=True),annot=True)
plt.show()
```

```python
#Cambiando mapa de colores
sns.heatmap(tips.corr(numeric_only=True),annot=True,cmap='coolwarm')
plt.show()
```

```python
#Separa un poco las l√≠neas
sns.heatmap(tips.corr(numeric_only=True),annot=True,cmap='coolwarm',linewidths=5)
plt.show()
```

```python
#colores a las l√≠neas
sns.heatmap(tips.corr(numeric_only=True),annot=True,cmap='coolwarm',linewidths=5,linecolor='black')
plt.show()
```

```python
#establecer m√≠nimos y m√°ximos para la correlacion
sns.heatmap(tips.corr(numeric_only=True),annot=True,cmap='coolwarm',linewidths=5,linecolor='black',
            vmin=0.5,vmax=1)
plt.show()
```

```python
#quitar la barra lateral
sns.heatmap(tips.corr(numeric_only=True),annot=True,cmap='coolwarm',linewidths=5,linecolor='black',
            vmin=0.5,vmax=1,cbar=False)
plt.show()
```
