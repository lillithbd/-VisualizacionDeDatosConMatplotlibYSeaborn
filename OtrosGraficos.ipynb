### Importar librerías

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Cargando Datos

```python
data = np.random.randint(1,50,100)
```

### Histograma

```python
plt.hist(data)
plt.show()
```

### Número de barras con bins

```python
plt.hist(data, bins=50) #número de barras en la cual llega la distribución de datos
plt.show()
```

### Cambiando estilo de las barras 

```python
plt.hist(data, bins=10, histtype='step') #número de barras en la cual llega la distribución de datos
plt.show()
```

### Grafico tipo BoxPlot

```python
plt.boxplot(data)
plt.show()
```

### Cambiar dirección, rellenar el rango inter cuartil patch_artist y enfocar la mediana notch

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
```

### Agregando outliers para manejarlo con la gráfica

```python
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=True)
plt.show()
```

### Grafico de disperson o Scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y)
plt.show()
```

### Scatter con área

```python
plt.scatter(x,y, s=area)
plt.show()
```

### Scatter con colores

```python
plt.scatter(x,y, s=area, c=colors)
plt.show()
```

### Scatter con marcadores

```python
plt.scatter(x,y, s=area, c=colors, marker='x')
plt.show()
```

```python
plt.scatter(x,y, s=area, c=colors, marker='o')
plt.show()
```

```python
plt.scatter(x,y, s=area, c=colors, marker='o', alpha=0.5)
plt.show()
```
