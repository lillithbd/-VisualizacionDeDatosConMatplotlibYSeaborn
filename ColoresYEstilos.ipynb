### Importar librerías

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Crear Set de Datos

```python
x = np.linspace(0,5,11)
```

### Agregar colores y estilos en matploblib

```python
print(plt.style.available) #Estilos de gráficas que hay
```

```python
plt.style.use('bmh')
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,'r')
ax.plot(x,x+2,'b')
ax.plot(x,x+3,'y')
ax.plot(x,x+4,'g')
```

### Personalizar colores y estilos en matlab

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'b-')
ax.plot(x,x+3,'y.-')
ax.plot(x,x+4,'go:')
```

### Personalizar colores y estilos en Pyplot

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,color = 'green')
ax.plot(x,x+2,color = 'blue')
ax.plot(x,x+3,color = '#66FF66')
ax.plot(x,x+4,color = '#C42152')
```

### Transparencia y ancho de línea

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,color = 'green', alpha=0.9, linewidth=5 )
ax.plot(x,x+2,color = 'blue',  linewidth=8)
ax.plot(x,x+3,color = '#66FF66',  linewidth=3)
ax.plot(x,x+4,color = '#C42152',  linewidth=2)
```

### Estilo de Línea

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,color = 'green', alpha=0.9, linewidth=5, linestyle='-', )
ax.plot(x,x+2,color = 'blue',  linewidth=8, linestyle='--')
ax.plot(x,x+3,color = '#66FF66',  linewidth=3, linestyle='dashed')
ax.plot(x,x+4,color = '#C42152',  linewidth=2, linestyle=':')
```

### Marcadores

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,color = 'green', alpha=0.9, linewidth=5, linestyle='-', marker='o')
ax.plot(x,x+2,color = 'blue',  linewidth=8, linestyle='--', marker='x')
ax.plot(x,x+3,color = 'yellow',  linewidth=3, linestyle='dashed', marker='v')
ax.plot(x,x+4,color = 'red',  linewidth=2, linestyle=':', marker='P')
```

### Cambiar tamaño a marker

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,color = 'green', alpha=0.9, linewidth=5, linestyle='-', marker='o', markersize=8)
ax.plot(x,x+2,color = 'blue',  linewidth=8, linestyle='--', marker='x')
ax.plot(x,x+3,color = 'yellow',  linewidth=3, linestyle='dashed', marker='v')
ax.plot(x,x+4,color = 'red',  linewidth=2, linestyle=':', marker='P')
```

### Cambiar color al marcador

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,color = 'green', alpha=0.9, linewidth=5, linestyle='-', marker='o', markersize=8, markerfacecolor='yellow')
ax.plot(x,x+2,color = 'blue',  linewidth=8, linestyle='--', marker='x')
ax.plot(x,x+3,color = 'yellow',  linewidth=3, linestyle='dashed', marker='v')
ax.plot(x,x+4,color = 'red',  linewidth=2, linestyle=':', marker='P')
```
