### Importando librerías

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Cargando datos

```python
tips = sns.load_dataset('tips')
tips
```

### Creando tipo gráfica con displot

```python
sns.displot(data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip') #automáticamente optimiza la gráfica según los datos que se tengan
plt.show()
```

### Agrupamiento por otra variable

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex') #si le quito el eje y crea un histograma
plt.show()
```

### Especificar el tipo de gráfico con el que quiero trabajar con kind

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde') 
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=True) 
plt.show()
```

### Cambiar tonalidades de colores

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=True, palette='dark') 
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=True, palette='dark', alpha=.25) 
plt.show()
```
