### Importando librer√≠as

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Cargando datos

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

### JointPlot

```python
sns.jointplot(data=tips, x='total_bill', y= 'tip');
plt.show()
```

```python
#Segmentarlo por la variable sexo
sns.jointplot(data=tips, x='total_bill', y= 'tip',hue='sex');
plt.show()
```

```python
#Especificando el diagrama que queremos
sns.jointplot(data=tips, x='total_bill', y= 'tip',hue='sex',kind='hist');
plt.show()
```

```python
#Especificando el diagrama que queremos
sns.jointplot(data=tips, x='total_bill', y= 'tip',hue='sex',kind='kde');
plt.show()
```

```python
#volviendo al histograma y usando marginal:ticks 
sns.jointplot(data=tips, x='total_bill', y= 'tip',hue='sex',kind='hist',marginal_ticks=True);
plt.show()
```

```python
#Usando keywords
sns.jointplot(data=tips, x='total_bill', y= 'tip',hue='sex',kind='hist',marginal_ticks=True,marginal_kws=dict(bins=25,fill=False,multiple='dodge'));
plt.show()
```

### Pairplot

```python
sns.pairplot(data=tips);
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex');
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm',corner=True);#para no tener info duplicada
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm',corner=True, kind='scatter');
plt.show()
```
