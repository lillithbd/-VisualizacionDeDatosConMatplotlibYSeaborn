### Importando librerías

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Cargando datos

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

### Gráfica de cantidad

```python
sns.countplot(data = tips, x= 'day', hue='sex'); 
plt.show()
```

### Gráfico stripplot 

```python
sns.stripplot(data = tips, x= 'day', hue='sex'); 
plt.show()
```

```python
sns.stripplot(data = tips, x= 'day', y='total_bill', hue='sex'); 
plt.show()
```

```python
sns.stripplot(data = tips, x= 'day', y='total_bill', hue='sex', dodge=True); 
plt.show()
```

### Gráfico Swarmplot

```python
sns.swarmplot(data = tips, x= 'day', y='total_bill', hue='sex', dodge=True); 
plt.show()
```

### Gráfico Boxplot


```python
sns.boxplot(data = tips, x= 'day', y='total_bill', hue='sex', dodge=True); 
```

### Combinar gráficos

```python
sns.boxplot(data = tips, x= 'day', y='total_bill', hue='sex', dodge=True); 
sns.swarmplot(data = tips, x= 'day', y='total_bill', hue='sex', dodge=True, color="0"); 
plt.show()
```

```python
sns.boxplot(data = tips, x= 'day', y='total_bill', hue='sex', dodge=True); 
sns.swarmplot(data = tips, x= 'day', y='total_bill', hue='sex', dodge=True, color="0", marker='<'); 
plt.show()
```

### Gráfico violinplot

```python
sns.violinplot(data = tips, x= 'day', y='total_bill', hue='sex', dodge=True); 
plt.show()
```

```python
sns.violinplot(data = tips, x= 'day', y='total_bill', hue='sex', palette='pastel'); 
plt.show()
```

```python
sns.violinplot(data = tips, x= 'day', y='total_bill', hue='sex', split=True,palette='pastel'); 
plt.show()
```

```python
sns.violinplot(data = tips, y='total_bill'); 
plt.show()
```

### Grafico catplot

```python
sns.catplot(data = tips, x= 'day', y='total_bill', hue='sex', dodge=True); 
plt.show() #él mismo define que es mejor utilizar con estos datos
```

```python
sns.catplot(data = tips, x= 'day', y='total_bill', hue='sex', dodge=True,kind='swarm'); #le puedo decir el tipo de gráfico
plt.show() 
```

```python
#Especificar una columna adicional para clasificar los datos
sns.catplot(data = tips, x= 'day', y='total_bill', hue='sex', dodge=True,kind='box', col='time');
plt.show() 
```
